```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

### Basic Graph

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# resizing graph - (dpi specifies pixels per inch.  When saving should use 300 if possible)
plt.figure(figsize=(5,3), dpi=200)

#plt.plot([1, 2, 3], [2, 4, 6], label = '2x', color = 'red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

# Using shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--', label='2x')

## line number two
x2=np.arange(0,4.5,0.5)

plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')
plt.plot(x2[5:], x2[5:]**2, 'r--')

plt.title('Basic Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

## change ticks using yticks and xticks
#plt.xticks([1, 2, 3])
#plt.yticks([2, 4, 6])

# show legend
plt.legend()

# can also save by using
# plt.savefig('xxxxx')

# show plot
plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars=plt.bar(labels, values)

# add character to the bars
bars[0].set_hatch('/')
bars[1].set_hatch('*')
bars[2].set_hatch('|')
# 


# resize
plt.figure(figsize=(6, 4))

# show
plt.show()
```
